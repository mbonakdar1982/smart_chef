```python
#Determine which of the componnets are in enough ingredients 
ing_valid = ingredients.groupby('component').size().sort_values()
select_from_ing = list(ing_valid[ing_valid.values > 100].index)
sorted(select_from_ing)
```

```python
max_amount = max(ingredients[ingredients['component'] == 'Calcium, Ca']['amount'])
max_amount
```

```python
selected_ingredients = ingredients[(ingredients['component'] == 'Calcium, Ca') & (ingredients['amount'] > max_amount / 5)].sort_values('amount', ascending=False)
selected_ingredients
```

```python
nlp = spacy.load("en_core_web_sm")
expanded_list = []
for item in selected_ingredients['description_food'].unique():
    expanded_list.extend(re.split(';|,| ', item))

new_list = set([i.strip().lower() for i in expanded_list if (not re.findall('[0-9"/\?><)\*\(\!\@#$\-%^&*]', i) and len(i.strip()) > 2 and nlp(i.strip())[0].pos_ == 'NOUN')])
new_list
```

```python
nlp = spacy.load("en_core_web_sm")
expanded_list = []
for item in selected_ingredients['description_food'].unique():
    expanded_list.extend([str(word).lower() for word in nlp(item) if word.pos_ == 'NOUN' and len(str(word)) > 2 and not re.findall('[0-9"/\?><)\*\(\!\@#$\-%^&*]', str(word))])
new_list = sorted(list(set(expanded_list)))
new_list
```
